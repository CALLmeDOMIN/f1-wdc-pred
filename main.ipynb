```python
import fastf1
import pandas as pd
```

```python
session = fastf1.get_session(2016, 5, 'R')

session.load(laps=True)
```

```python
driver_stats = []


for result in session.results.iterrows():
    driver = result[1]["Abbreviation"]
    
    avg_lap_time = session.laps.pick_drivers(driver)["LapTime"].mean()
    
    driver_stats.append({
        "Driver": result[1]["Abbreviation"],
        "TeamId": result[1]["TeamId"],
        "FinishPosition": result[1]["Position"],
        "PositionsGained": result[1]["Position"] - result[1]["GridPosition"],
        "PointsGained": result[1]["Points"],
        "AvgLapTime": avg_lap_time,
    })
    
driver_stats_df = pd.DataFrame(driver_stats)

driver_stats_df
```

```python
driver_stats_df = driver_stats_df.dropna()

driver_stats_df
```
